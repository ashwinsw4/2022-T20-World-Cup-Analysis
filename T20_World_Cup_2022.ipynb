Replacing missing values (-) with  0s

```python
import numpy as np
dfbat= pd.DataFrame(dfbat.replace('-', np.nan))
dfbat.fillna(0,inplace=True)
```

```python
dfbowl=pd.DataFrame(dfbowl.replace('-',np.nan))
dfbowl.fillna(0,inplace=True)
```

```python
dfbat.sum().isna()
```

Batters' impact in the tournament

```python
filter=dfbat[dfbat['Inns']>3]
```

```python
sorted_df = filter.sort_values(by='Batting Impact', ascending=False)

# Plot the top players using Plotly
fig = px.bar(
    sorted_df,
    x='Player',
    y='Batting Impact',
    labels={'Impact Score': 'Impact Score'},
    title='Players with Max Impact (Minimum 4 innings)',
    color=sorted_df['Batting Impact']
)

fig.update_layout(
    xaxis_title='Player',
    yaxis_title='Batting Impact',
    xaxis_tickangle=-45,
    showlegend=False,
)

fig.show()
```

Bowlers' impact in the tournament


```python
filters=dfbowl[dfbowl['Overs']>4]
sort=filters.sort_values(by='Bowling Impact',ascending=False)
figu=px.bar(sort,sort['Player'],sort['Bowling Impact'],color=sort['Bowling Impact'])
figu.update_layout(
    xaxis_title='Player',
    yaxis_title='Batting Impact',
    xaxis_tickangle=-45,
    showlegend=False,
)

figu.show()
```

Average runs scored by players for their team. (England 88.11 means every England batter on an average made 88 runs in the whole tournament)

```python
total_runs_by_team = dfbat.groupby('Country')['Runs'].mean().reset_index()

# Create a bar chart using Plotly Express
fig = px.bar(total_runs_by_team, x='Country', y='Runs', text='Runs', title='Average Runs By Batters from each Country',
             labels={'Runs': 'Total Runs', 'Country': 'Team'},color=total_runs_by_team['Runs'])

# Customize the layout
fig.update_layout(xaxis_title='Team', yaxis_title='Runs', title_x=0.5)
fig.update_layout(
    plot_bgcolor='black',  # Set the plot background color to black
    paper_bgcolor='black',  # Set the paper (outside the plot area) background color to black
    font=dict(color='white')  # Set font color to white for better visibility on a dark background
)

# Show the plot
fig.show()
```

In a batting dominated game SCG was the only stadium to give bowlers an even chance, where batting in 2nd innings was harder compared to other stadiums

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=df["venue"],
    y=df["first innings score"],
    name='First Innings Runs',
    marker_color='blue'
))
fig.add_trace(go.Bar(
    x=df["venue"],
    y=df["second innings score"],
    name='Second Innings Runs',
    marker_color='red'
))
fig.update_layout(barmode='group',
                  xaxis_tickangle=-45,
                  title="Best Stadiums to Bat First or Chase")
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=dfbat["Country"],
    y=dfbat["6s"],
    name='6s scored',
    marker_color='blue'
))
fig.add_trace(go.Bar(
    x=dfbat["Country"],
    y=dfbat["4s"],
    name='4s scored',
    marker_color='red'
))
fig.update_layout(barmode='group',
                  xaxis_tickangle=-45,
                  title="Comparison of  4s and 6s")
fig.show()
```

```python


dfpakbat=dfbat[dfbat['Country']=='PAK']

dfengbat=dfbat[dfbat['Country']=='ENG']

dfpakbowl=dfbowl[dfbowl['Country']=='PAK']

dfengbowl=dfbowl[dfbowl['Country']=='ENG']

```

Comparision between the finalists

```python
import matplotlib.pyplot as plt


average_score_england = dfengbat['Runs'].mean()
average_score_pakistan = dfpakbat['Runs'].mean()


total_runs_england = dfengbat['Runs'].sum()
total_runs_pakistan = dfpakbat['Runs'].sum()


fig = px.bar(x=['England', 'Pakistan'], y=[average_score_england, average_score_pakistan],
             labels={'x': 'Country', 'y': 'Average Score'},
             title='Average Scores Comparison')


fig.show()


fig = px.bar(x=['England', 'Pakistan'], y=[total_runs_england, total_runs_pakistan],
             labels={'x': 'Country', 'y': 'Total Runs'},
             title='Total Runs Comparison')


fig.show()
```

```python
import matplotlib.pyplot as plt


average_wkts_england = dfengbowl['Wkts'].mean()
average_wkts_pakistan = dfpakbowl['Wkts'].mean()


total_wkts_england = dfengbowl['Wkts'].sum()
total_wkts_pakistan = dfpakbowl['Wkts'].sum()


fig = px.bar(x=['England', 'Pakistan'], y=[average_wkts_england, average_wkts_pakistan],
             labels={'x': 'Country', 'y': 'Average Wkts'},
             title='Average Wkts Comparison')


fig.show()


fig = px.bar(x=['England', 'Pakistan'], y=[total_wkts_england, total_wkts_pakistan],
             labels={'x': 'Country', 'y': 'Total Wkts'},
             title='Total Wkts Comparison')


fig.show()
```
